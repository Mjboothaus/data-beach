```python
import pandas as pd
```

```python
data_URL = "https://raw.githubusercontent.com/DataBooth/beach-safe/main/data/beach_data.csv"
```

```python
beach_df = pd.read_csv(data_URL)
```

```python
beach_df.head()
```

```python
from_address = '42 Wrights Road Drummoyne Australia'
region = 'AU'
```

```python
to_address = "Gravelly Beach"
```

```python
def calculate_time_between_locations(from_address, to_address):
  route_time = None
  route_dist = None
  try:
    route = WazeRouteCalculator.WazeRouteCalculator(from_address, to_address, region)
    route_time, route_dist = route.calc_route_info()
  except Exception as e:
    print(e)
  return route_time, route_dist
```

```python
time, dist = calculate_time_between_locations(from_address, to_address)
```

```python
print(time, dist)
```

```python
beach_df['travel_time_dist'] = beach_df['Site'].apply(lambda site: calculate_time_between_locations(from_address, site))
```

```python
beach_df
```

```python
def get_first_element(time_dist):
  if time_dist != None:
    return time_dist[0]
  else:
    return None
```

```python
def get_second_element(time_dist):
  if time_dist != None:
    return time_dist[1]
  else:
    return None
```

```python
beach_df['time_in_minutes'] = beach_df['travel_time'].apply(lambda time_dist: get_first_element(time_dist))
```

```python
beach_df['time_in_hours'] = beach_df['time_in_minutes'] / 60.0
```

```python
beach_df['dist_in_kms'] = beach_df['travel_time'].apply(lambda time_dist: get_second_element(time_dist))
```
